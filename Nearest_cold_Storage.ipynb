```python
import gradio as gr
import pandas as pd
from tabulate import tabulate

df = x  # Assuming 'x' is the preloaded DataFrame

# List of categories
categories = [
    "F&V (F&V, Spices)", "Marine & Fishery", "Dairy", "Irradiation", "Fishery/Marine",
    "Marine", "Meat", "F&V", "RTE/RTC", "RTE", "Mixed (Dairy, F&V)", "Mixed (Dairy, Meat, F&V)",
    "Mixed (Meat, Poultry & Dairy)", "Meat & Poultry", "F&V (Pulses, F&V, Spices)",
    "Mixed (F&V, Meat & Fishery)", "Poultry"
]

def get_projects(state, district):
    # Filter data based on inputs
    filtered_projects = df[df["District"] == district]

    if filtered_projects.empty:
        return "No data found for the given filters."

    # Convert DataFrame to a clean table format (without index numbers)
    table = tabulate(
        filtered_projects[['Project', 'Sector', 'District', 'State','Emails']],
        headers=["ColdStorage", "Type of Storage", "District", "State","Emails"],  # Clean column names
        tablefmt="grid",
        showindex=False  # Remove index numbers
    )

    return f"<pre>{table}</pre>"

def get_State_projects(state, district):
    # Filter data based on inputs
    filtered_projects = df[(df["State"] == state)]

    if filtered_projects.empty:
        return "No data found for the given filters."

    # Convert DataFrame to a clean table format (without index numbers)
    table = tabulate(
        filtered_projects[['Project', 'Sector', 'District', 'State','Emails']],
        headers=["ColdStorage", "Type of Storage", "District", "State","Emails"],  # Clean column names
        tablefmt="grid",
        showindex=False  # Remove index numbers
    )

    return f"<pre>{table}</pre>"

with gr.Blocks() as iface:
    with gr.Tabs():
        # with gr.TabItem("Home"):
        #     gr.Markdown("# Welcome to Cold Storage Predictor")

        with gr.TabItem("Predict Nearest Cold Storage"):
            gr.Markdown("## AGRICOLD")

            # Input fields
            state_input = gr.Textbox(label="Enter State Name")
            district_input = gr.Textbox(label="Enter District Name")

            # Predict button
            predict_button = gr.Button("Predict districtwise Nearest Cold Storage")
            output = gr.HTML()

            predict_button.click(
                fn=get_projects,
                inputs=[ state_input, district_input],
                outputs=output
            )

            predict_button = gr.Button("Predict statewise Nearest Cold Storage")
            output = gr.HTML()

            predict_button.click(
                fn=get_State_projects,
                inputs=[ state_input, district_input],
                outputs=output
            )

        # with gr.TabItem("Contact"):
        #     gr.Markdown("## Contact Us\nFor inquiries, reach out to us!")

iface.launch(server_name="0.0.0.0", server_port=3030,share=True)
```
